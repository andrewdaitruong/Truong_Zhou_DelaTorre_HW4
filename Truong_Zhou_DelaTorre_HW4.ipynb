# 1. Consider the two tables shown below-called population and countyseats.


```python
import pandas as pd
```

```python
population = pd.DataFrame (
   { "state": ["California", "California", "California", "California"],
    "county": ["Orange", "Orange", "Los Angeles", "Los Angeles"],
    "year": [2000, 2010, 2000, 2010],
    "population": [2846289, 3010232, 3694820, 3792621]
   }
)
```

```python
population.index = range(1, len(population) + 1)
population
```

```python
countyseats = pd.DataFrame (
    { "statename": ["California", "California", "California", "Oregon"],
    "countyname": ["Orange", "Los Angeles", "San Diego", "Wasco"],
    "countyseat": ["Santa Ana", "Los Angeles", "San Diego", "The Dalles"]
    }
)
```

```python
countyseats.index = range(1, len(countyseats) + 1)
countyseats
```

```python
countyseats_rename = countyseats.copy()
countyseats_rename = countyseats_rename.rename(columns={"countyname": "county", "statename": "state"})
```

```python
pd.merge(population, countyseats_rename)
```

# 2.

```python
df = pd.read_csv('billboard.csv', encoding='ISO-8859-1')
df
```

```python
week_columns = [col for col in df.columns if 'week' in col]
tidy_df = df.melt(id_vars=["year", "artist.inverted", "track", "time", "genre", "date.entered", "date.peaked"],
                  value_vars=week_columns,
                  var_name="week", value_name="rank")
```

```python
tidy_df = tidy_df.dropna(subset=["rank"])
tidy_df.reset_index(drop=True, inplace=True)
tidy_df
```

# a) Show for each track, how many weeks it spent on the chart

```python
weeks = tidy_df.groupby("track")["week"].nunique()
weeks
```

# b) List tracks in decreasing order of number of weeks spent on the chart

```python
weeks_descend = weeks.sort_values(ascending=False)
weeks_descend
```

# c) Show for each track, its top rank

```python
top_rank = tidy_df.groupby("track")["rank"].min()
top_rank
```

# d) List tracks in increasing order of its top rank

```python
top_rank_ascend = top_rank.sort_values()
top_rank_ascend
```

# e) Show for each artist, their top rank

```python
top_rank_artist = tidy_df.groupby("artist.inverted")["rank"].min()
top_rank_artist
```

# f) List artists in increasing order of their top rank

```python
top_artist_ascend = top_rank_artist.sort_values(ascending=False)
top_artist_ascend
```

# g) List tracks that spent more than 35 weeks in the charts

```python
weeks_spent = tidy_df.groupby("track")["week"].nunique()
tracks_over_35_weeks = weeks_spent[weeks_spent > 35]
tracks_over_35_weeks
```

# h) List tracks that spent more than 35 weeks in the charts along with their artists

```python
weeks_artist = tidy_df.groupby(["track", "artist.inverted"])["week"].nunique()
tracks_and_artists_over_35_weeks = weeks_artist[weeks_artist > 35]
tracks_and_artists_over_35_weeks
```

# (a) The data is not “tidy”. In 2-3 sentences, explain why.

One reason is because there are multiple observations in a single row (Ex: row 3 observes both population ages 80+ and female). The column "Series Name" also has many data (Ex: population, sex, etc) which should ideally be spread in multiple columns.

# (b) Transform the table to tidy data with one country per row. [Give code]

```python
tidy_demo = demo.pivot_table(index=["Country Name", "Country Code"], columns="Series Code", values="YR2015", aggfunc="first").reset_index()
tidy_demo
```

# (c) Add the male/female population numbers together (i.e., ignore sex-related differences).

```python
tidy_demo.columns
```

```python
column_pairs = [
    ("SP.DYN.AMRT.FE", "SP.DYN.AMRT.MA"),
    ("SP.POP.0014.FE.IN", "SP.POP.0014.MA.IN"),
    ("SP.POP.1564.FE.IN", "SP.POP.1564.MA.IN"),
    ("SP.POP.65UP.FE.IN", "SP.POP.65UP.MA.IN"),
    ("SP.POP.80UP.FE", "SP.POP.80UP.MA")
]
```

```python
fe_columns = [col for col in tidy_demo.columns if '.FE' in col]
ma_columns = [col.replace('.FE', '.MA') for col in fe_columns if col.replace('.FE', '.MA') in tidy_demo.columns]
```

```python
for fe_col, ma_col in zip(fe_columns, ma_columns):
    new_col = fe_col.replace('.FE', '')
    fe_values = pd.to_numeric(tidy_demo[fe_col], errors="coerce").astype(float)
    ma_values = pd.to_numeric(tidy_demo[ma_col], errors="coerce").astype(float)
    tidy_demo[new_col] = fe_values + ma_values
```

```python
tidy_demo = tidy_demo.drop(columns=fe_columns + ma_columns)
pd.set_option('display.float_format', lambda x: '%.15g' % x)
tidy_demo
```

# d) Write code to show the top 5 countries with the lowest proportion of the population below 14 years old [Code, and list of 5 countries]

```python
tidy_demo['SP.POP.0014.IN'] = pd.to_numeric(tidy_demo['SP.POP.0014.IN'], errors='coerce')
tidy_demo['SP.POP.TOTL'] = pd.to_numeric(tidy_demo['SP.POP.TOTL'], errors='coerce')

lowest_proportion = tidy_demo[['Country Name', 'SP.POP.0014.IN', 'SP.POP.TOTL']].copy()

lowest_proportion.loc[:, 'proportion_below_14'] = lowest_proportion['SP.POP.0014.IN'] / lowest_proportion['SP.POP.TOTL']

lowest_proportion_sorted = lowest_proportion.sort_values(by="proportion_below_14").head(5)

lowest_proportion_sorted[['Country Name', 'proportion_below_14']]
```
